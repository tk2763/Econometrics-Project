```python
#重回帰分析
#参照：https://qiita.com/karaage0703/items/f38d18afc1569fcc0418

import pandas as pd
import numpy as np

df = pd.read_csv('aggregate_startups_ADD.csv')

df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.datasets import load_boston
import statsmodels.api as sm

#https://dataanablog.com/statistical-methods-in-python-for-beginners-multiple-regression-analysis/
```

```python
y1 = df["logstartups"]
X1 = df[["treated_post","treated","post"]]

X1_con = sm.add_constant(X1)
```

```python
model1 = sm.OLS(y1, X1_con)  # モデルを作成

result1 = model1.fit()      # モデルを適応
```
